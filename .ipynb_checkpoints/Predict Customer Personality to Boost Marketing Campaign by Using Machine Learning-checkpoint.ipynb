```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import datetime as dt
from datetime import date
```

```python
df = pd.read_csv('marketing_campaign_data.csv')
df.head()
```

```python
df.info()
```

```python
df.drop(['Unnamed: 0'], inplace=True, axis=1)
```

```python
df.describe()
```

```python
df['Dt_Customer'] = df['Dt_Customer'].apply(lambda x : x.split('-')[-1])

df['Age'] = np.int64(df['Dt_Customer']) - df['Year_Birth']

df['Age Category'] = df['Age'].apply(lambda x : 'Adult' 
                                     if x < 40 else ('Middle Aged' 
                                                     if x < 60 else ('Senior')))
```

```python
dfgroup  = df[df['Response'] == 0].groupby(['Age Category']).aggregate({'ID':'count'}).reset_index()
dfgroup1 = df[df['Response'] == 1].groupby(['Age Category']).aggregate({'ID':'count'}).reset_index()

dfgroup1['Conversion rate'] = dfgroup1['ID'] / (dfgroup1['ID'] + dfgroup['ID']) * 100

dfgroup1['Age Category'] = pd.Categorical(dfgroup1['Age Category'], categories=['Adult', 'Middle Aged', 'Senior'])
dfgroup1 = dfgroup1.sort_values('Age Category')
dfgroup1
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(x = dfgroup1['Age Category'],
                 y = 'Conversion rate', 
                 data = dfgroup1, 
                 palette='tab10')

plt.title('Conversion Rate (Percentage) Based on Age Category', weight='bold', fontsize = 20)
sns.despine()
plt.tight_layout()

plt.bar_label(ax.containers[0], fmt='%.2f%%')
plt.show()
```

## Data Cleansing & Pre-Processing

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df = df.drop(columns='ID')
df.duplicated().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencod = LabelEncoder()

df['Marital_Status'] = labelencod.fit_transform(df['Marital_Status'])
df['Age Category'] = labelencod.fit_transform(df['Age Category'])
df['Education'] = labelencod.fit_transform(df['Education'])
```

```python
from sklearn.preprocessing import StandardScaler
standscale = StandardScaler()

df['Income'] = standscale.fit_transform(df[['Income']])
df['MntCoke'] = standscale.fit_transform(df[['MntCoke']])
df['MntFruits'] = standscale.fit_transform(df[['MntFruits']])
df['MntMeatProducts'] = standscale.fit_transform(df[['MntMeatProducts']])
df['MntFishProducts'] = standscale.fit_transform(df[['MntFishProducts']])
df['MntSweetProducts'] = standscale.fit_transform(df[['MntSweetProducts']])
df['MntGoldProds'] = standscale.fit_transform(df[['MntGoldProds']])
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(df.corr(), annot= True)
```

```python
df['Child'] = df['Kidhome'] + df['Teenhome']
df['Total'] = df['MntCoke'] + df['MntFruits'] + df['MntMeatProducts'] + df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df['Total_Purchase'] = df['NumWebPurchases'] + df['NumCatalogPurchases'] + df['NumStorePurchases']
```

```python
df.head()
```

## Modeling

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)



plt.figure(figsize=(10, 5))
plt.title('Inertia Evaluation Score')
sns.lineplot(x=range(2, 11), y=inertia, color='blue', linewidth = 3)
sns.scatterplot(x=range(2, 11), y=inertia, s=200, color='red',  linestyle='-')
plt.xlabel('n_clusters')
plt.show()
```

```python
(pd.Series(inertia) - pd.Series(inertia).shift(-1)) / pd.Series(inertia) * 100
```

```python
from sklearn.metrics import silhouette_score

range_n_clusters = list(range(2,11))
arr_silhouette_score_euclidean = []
for i in range_n_clusters:
    kmeans = KMeans(n_clusters=i).fit(df)
    preds = kmeans.predict(df)
    score_euclidean = silhouette_score(df, preds, metric='euclidean')
    arr_silhouette_score_euclidean.append(score_euclidean)

fig, ax = plt.subplots(figsize=(8, 6))
plt.title('Silhouette Evaluation Score')
sns.lineplot(x=range(2,11), y=arr_silhouette_score_euclidean, color='blue', linewidth = 3)
sns.scatterplot(x=range(2,11), y=arr_silhouette_score_euclidean, s=200, color='red',  linestyle='-')
plt.xlabel('n_clusters')
plt.show()
```

```python
clast =arr_silhouette_score_euclidean.index(max(arr_silhouette_score_euclidean)) + 2
```

```python
kmeans = KMeans(n_clusters=clast, random_state=42).fit(df)
clusters = kmeans.labels_
df['clusters'] = clusters
```

```python
from sklearn.decomposition import PCA 

pca = PCA(n_components=3)
pca.fit(df)
pcs = pca.transform(df)
final_pca = pd.DataFrame(data = pcs, columns = ['PC1', 'PC2','PC3'])
final_pca['clusters'] = clusters
final_pca.head()
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.scatterplot(x='PC1', y='PC2',hue='clusters',palette = 'tab10', linestyle= '-', s=100,data=final_pca,ax=ax)
plt.show()
```
